(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
# write your answer here
arr_elems_Comn = [ ]
elements = input("Enter array elements\t").split(',') # taking array elements from user
arr1_Int = [int (i) for i in elements] #converting string elements into integer
for i in range(0,len(arr1_Int)):
  for j in range(i+1,len(arr1_Int)):
    if arr1_Int[i] == arr1_Int[j]:
      arr_elems_Comn.append(arr1_Int[j])
print("elements common in array is: {}".format(arr_elems_Comn))
if (sum(arr_elems_Comn) == 0):
  print(-1)
```

(2) Write a Python program to select all the Sundays of a specified year. (4 points)



```python
# write your answer here
# import datetime as dt
# x = dt.datetime.now()
# print(x.month)
# print(x.strftime("%A"))
# def date(x):
import datetime as dt 
import pandas as pd
All_Dates = [ ]
All_Months = [ ]
All_Sunday = [ ]
All_Year = [ ]
year = int (input("enter year:\t"))
O = dt.date(year, 1, 1)
O = O + dt.timedelta(days= 6 - O.weekday())
while (O.year == year):
    All_Months.append(O.strftime("%B"))
    All_Dates.append(O)
    All_Sunday.append(O.strftime("%A"))
    All_Year.append(O.year)
    O = O + dt.timedelta(days=7)
columns = ['date','year','month','day']
df = pd.DataFrame({'Date':All_Dates,'Month': All_Months,'Day':All_Sunday,'Year': All_Year})
df
```
